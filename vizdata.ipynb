```python
import pandas as pd
import numpy as np
import scipy as sp
from scipy.stats import iqr
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('user1_grasp.csv')
df.describe()
df.hist(figsize =(20,10))
plt.show()
```

```python
d_mean = df.groupby('class').aggregate(np.mean)**2
d_mean = d_mean.apply(np.sqrt)
d_mean.index = ['unmarked data','hand at rest','clenched fist', 'wrist flexion', 'wrist extension','radial deviations','ulnar deviations']
d_mean.columns = ['class','channel1','channel2','channel3','channel4','channel5','channel6','channel7','channel8']
```

```python
d_mean.plot.bar(rot=0,figsize=(12,10))
plt.ylabel('Root Mean Square (Voltage)')
plt.xlabel('Hand Gestures')
plt.title('Root Mean Square (Voltage) of Channel Responses for Class (Hand Gestures)')
plt.show()
```

```python
d_std = df.groupby('class').aggregate(np.std)
d_std.index = ['unmarked data','hand at rest','clenched fist', 'wrist flexion', 'wrist extension','radial deviations','ulnar deviations']
d_std.columns = ['time','channel1','channel2','channel3','channel4','channel5','channel6','channel7','channel8']
d_std = d_std.drop(columns = 'time')

```

```python
d_std.plot.bar(rot=0,figsize=(12,10))
```

```python
d_kur = df.groupby('class').apply(pd.DataFrame.kurt)
d_kur = d_kur.drop(columns='time')
d_kur.plot.bar(rot=0,figsize=(12,10))
```
